```python
import gpxpy
import gpxpy.gpx
import copy
```

```python
gpx_file = open('./gpxs/001.gpx', 'r')

gpx = gpxpy.parse(gpx_file)

gpx_o = copy.deepcopy(gpx)
```

```python
gpx.remove_elevation()
gpx.remove_time()
```

```python
print(len(gpx_o.get_points_data()))

gpx.reduce_points(min_distance=50)

print(len(gpx.get_points_data()))
```

```python
gpx.to_xml()
```

```python

outfile = 'test_001.gpx'
with open(outfile, 'w') as f:
    f.write(gpx.to_xml())
```
